```python
#import library
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
#read photo
img = cv2.imread("/content/Input-test-images-a-Cameraman-grayscale-b-grayscale-Lena-and-c-color-Lena-All_Q640.jpg")

#Declaration

#Kernel
Kernel = np.ones((5,5), np.float32)/25

#Convolved Img
ConImg = cv2.filter2D(img, -1, Kernel)

#Blur Img
blurImg = cv2.blur(img,(5,5))

#Gaussian Blur Img
Gblur = cv2.GaussianBlur(img,(5,5), 0)
```

```python
#original image
plt.subplot(221), plt.imshow(img), plt.title('Original Image')
plt.xticks([]), plt.yticks([])

#convolved image
plt.subplot(222), plt.imshow(ConImg), plt.title('Convolved Image')
plt.xticks([]), plt.yticks([])

#blur image
plt.subplot(223), plt.imshow(blurImg), plt.title('Blur Image')
plt.xticks([]), plt.yticks([])

#GaussianBlur Image
plt.subplot(224), plt.imshow(Gblur), plt.title('GaussianBlur Image')
plt.xticks([]), plt.yticks([])

#show image
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
